```python
# Count even and odd numbers in a list

def even_count(mylist):
  count_even = 0
  for i in range(0, len(mylist)):
    if (mylist[i] % 2 == 0):
      count_even = count_even + 1
  return count_even

def odd_count(mylist):
  count_odd = 0
  for i in range(0, len(mylist)):
    if (mylist[i] % 2 != 0):
      count_odd = count_odd + 1
  return count_odd

mylist = [1,5,2,8,7,9]

E=even_count(mylist)
O=odd_count(mylist)

print(E)
print(O)
```

```python
# Count positive or negative numbers in a list

def positive(mylist):
  count_pos = 0
  for i in range(0, len(mylist)):
    if (mylist[i] >= 0):
      count_pos = count_pos + 1
  return count_pos

def negative(mylist):
  count_neg = 0
  for i in range(0, len(mylist)):
    if (mylist[i] < 0):
      count_neg = count_neg + 1
  return count_neg

mylist = [1, -2, 5, -9, -10]
P= positive(mylist)
N = negative(mylist)

print(P)
print(N)
```

```python
# reverse a list

def reverse_list(numlist):
  i = 0
  j = len(numlist) - 1

  while(i < j):
    temp = numlist[i]
    numlist[i] = numlist[j]
    numlist[j] = temp
    i = i + 1
    j = j - 1
  return numlist

numlist = [10,20,30,40,50]
reverse_list(numlist)
```

```python
# sort list in ascending order

mylist = [20,10,80,50,40,60]

def myfunction(mylist):
  for i in range(0, len(mylist)):
    for j in range(i+1, len(mylist)):
      if (mylist[i] > mylist[j]):
        temp = mylist[i]
        mylist[i] = mylist[j]
        mylist[j] = temp

  return mylist

myfunction(mylist)
```

```python
# find second largest number in a list

mylist = [20,50,40]

def myfunction(mylist):
  first = second = mylist[0]

  for i in range(1, len(mylist)):
    if (mylist[i] > first): 
      second = first     
      first = mylist[i]  
    elif (mylist[i] > second and mylist[i] < first):
      second = mylist[i]

  print("Largest number is : ", first)
  print("Second largest number is : ", second)

myfunction(mylist)
```
